```python
import os
import pandas as pd
```

```python
path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
```

```python
all_data = pd.read_csv("all_data_copy.csv")
all_data.head()
```

```python
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Month 2'] = pd.to_datetime(all_data['Order Date']).dt.month
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
```

```python
all_data.groupby(['Month']).sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,all_data.groupby(['Month']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

# The city with most number of sales 

```python
### Add city column
def get_city(address):
    return address.split(',')[1].strip()

def get_state(address):
    return str(address.split(',')[2])[1:3]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

```python
all_data.groupby(['City']).count()
```

```python
import matplotlib.pyplot as plt

cities = [city for city,df in all_data.groupby(['City'])]

plt.bar(cities,all_data.groupby(['City']).sum()['Sales'])
plt.xticks(cities, rotation ='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()
```

# At what time to show ad to maximize the likelihood of buying a product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data['Time'] = all_data['Order Date'].dt.hour
all_data
```

```python
result = all_data.groupby(['Time']).sum()
result.
```

```python
result.sort_values(['Sales'], ascending=False)
```

```python
hours = [hours for hours,df in all_data.groupby(['Time'])]

plt.plot(hours,all_data.groupby(['Time']).count()['Sales'])
plt.xticks(hours)
plt.grid()
plt.xlabel('Hours')
plt.ylabel('Orders amount')
plt.show()
```

# What products are most often sold together?

```python
all_data.groupby(['Order ID']).count().sort_values(['Product'])
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df
```

```python
from collections import Counter
from itertools import combinations

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

count.most_common(10)
```

# What product is saled the most?

```python
all_data.groupby('Product').sum()
```

```python
products = [product for product,df in all_data.groupby('Product')]

plt.bar(products, all_data.groupby('Product').sum()['Sales'])
plt.xticks(products, rotation ='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('Product')
plt.title('Total sales by product')
plt.show()
```

```python
plt.bar(products, all_data.groupby('Product').sum()['Quantity Ordered'])
plt.xticks(products, rotation ='vertical')
plt.ylabel('Number of sales')
plt.xlabel('Product')
plt.title('Total quantity saled by product')
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(products, all_data.groupby('Product').sum()['Quantity Ordered'])
ax2.plot(products, all_data.groupby('Product').sum()['Sales'],color='y')
ax1.set_xticklabels(products, rotation='vertical')
fig.show()
```
